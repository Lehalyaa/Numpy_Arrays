```python
# Declaration of Numpy

import numpy as np

a = np.array([10, 20, 30, 40])
b = np.array([[1, 2, 3], [4, 5, 6]])

print(a)
print(b)
```

```python
# Dimension of Array

print("Dimension of array A is: ", a.ndim)
print("Dimension of array  B is: ", b.ndim )
```

```python
# Shape of an Array

print("Dimension of A: ", a.shape)
print("Dimension of B: ", b.shape)
```

```python
# Data type of elements in an array

print("The data type of elements in A is", a.dtype)
print("The data type of elements in B is", b.dtype)
```

## In - Built Functions in NumPy

```python
# Printing a Zero Matrix

np.zeros((2,3))
```

```python
# Printing an Identity Matrix

np.ones((3,6))
```

```python
# Diagonal matrix

np.eye((3))
```

```python
# printing a range of range with Step Size

np.arange(1,11,2) # Same as range function, stop value is not included
```

```python
# dividing a number into specified terms

np.linspace(0,1,5) # 1 is also included
```

```python
# Arithmetic oprations in Arrays

c = a * 2
d = a + 5
print(c, "\n")
print (d)
```

```python
# Mean of elements in an array

print("The average of elements in A is: ",np.mean(a))
print("The average of elements in B is: ",np.mean(b))
```

```python
# Median of elements

print("Median of A: ",np.median(a))
print("Median of B: ", np.median(b))

''' For clean data mean and median are same, if different take the value of
median which is more appropriate'''
```

```python
# Aggregate Functions

print("Operations performed in Matrix A")
print("Max of A: ",np.max(a))
print("Min of A: ",np.min(a))
print("Sum of A: ",np.sum(a), "\n")


print("Operations Performed in Matrix B")
print("MAx of B: ", np.max(b))
print("Min of B: ", np.min(b))
print("Sum of B: ", np.sum(b))
```
